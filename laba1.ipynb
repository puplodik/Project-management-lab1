# Задание 1

```python
def convert_char_to_integer(char):
    if char.isdigit():
        return int(char)
    else:
        return None

char = input("Введите переменную строкового типа: ")
result = convert_char_to_integer(char)
if result is not None:
    print(f"Результат преобразования: {result}")
    print(f"Проверка: {result+5}")
```
